```python
import ROOT
from array import array
from ROOT import gROOT 
import numpy as np
```

```python
xlow, xhigh = 0.375, 0.725
d_func = ROOT.TF1('data_g', 'gaus(0)', xlow, xhigh)
d_func.SetParameters(1, 0.5384, 0.0481)
d_func.SetLineColor(4)
d_func.SetTitle("Gaussian Fits")
# d_func.SetLineStyle(2)
d_func.SetLineWidth(2)

g_func = ROOT.TF1('data_g', 'gaus(0)', xlow, xhigh)
g_func.SetParameters(1, 0.5492, 0.0500)
g_func.SetLineColor(30)
g_func.SetTitle("Gaussian Fits")
g_func.SetLineWidth(2)
# g_func.SetLineStyle(2)
```

```python

#Canvas to plot on and graph
c = ROOT.TCanvas()

#Draw canvas and graph
c.Draw()

l = ROOT.TLegend(0.7, 0.7, 0.9, 0.9)
l.AddEntry(d_func, "Data")
l.AddEntry(g_func, "GJets MC")

d_func.Draw()
g_func.Draw("SAME")
l.Draw("SAME")

etamass = 0.548
lheight = 1
#lheight=0.1
L = ROOT.TLine(etamass, 0, etamass, lheight)
L.SetLineColor(4)
L.SetLineStyle(2)
L.Draw("same")
```

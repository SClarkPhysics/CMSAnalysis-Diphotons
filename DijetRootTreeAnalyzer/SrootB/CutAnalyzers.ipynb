```python
import numpy as np
import pandas as pd
import os
```

```python
newdic = {"masym":[],
          "deta" :[],
          "dipho":[],
          "iso"  :[],
          "avg s/b":[],
          "max s/b":[],
          "min s/b":[],
          }

path = "CutOutFiles"
fcount = 0
for root, dirs, files in os.walk(path):
    for file in files:
        if("sorted" in file):
            thisDF = pd.read_csv(os.path.join(path,file))
            print("Reading file {}".format(os.path.join(path,file)))
            
            ms = thisDF["masym"].to_numpy()
            es = thisDF["deta"].to_numpy()
            ds = thisDF["dipho"].to_numpy()
            isos = thisDF["iso"].to_numpy()
            srbs = thisDF["srootb"].to_numpy()
            max_srb = srbs[0]
            min_srb = srbs[-1]
            avg_srb = np.mean(srbs)
            sum_srb = sum(srbs)
            
            newdic["masym"].append(ms[0])
            newdic["deta"].append(es[0])
            newdic["dipho"].append(ds[0])
            newdic["iso"].append(isos[0])
            newdic["avg s/b"].append(avg_srb)
            newdic["max s/b"].append(max_srb)
            newdic["min s/b"].append(min_srb)

#             newdic["sum s/b"].append(sum_srb)
            
            fcount += 1
#             if(fcount >= 1):break
            
            
```

```python
odframe = pd.DataFrame.from_dict(newdic)
dframe = odframe[['masym','deta','dipho','iso','avg s/b','max s/b','min s/b']]
# dframe = dframe[dframe.deta < 2.5]
dframe = dframe[dframe.deta>0].sort_values(by=["avg s/b"], ascending=False).reset_index(drop=True)
dframe.head(40)
```
